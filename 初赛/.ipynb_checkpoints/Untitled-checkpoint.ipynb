```python
import requests
import pandas as pd
prom_url = "http://47.102.202.42:31090/api/v1/query_range"
# metric_name = 'container_memory_working_set_bytes'
query = 'sum by (instance)(rate(node_cpu_seconds_total[%s])) * 100'%('5m')
metric_step = 86
response = requests.get(prom_url,
                        params={'query': query,
                                'start': 1658024433.419,
                                'end': 1658046033.419,
                                'step': metric_step})
results = response.json()['data']['result']
results
```

```python
latency_df = pd.DataFrame()
for result in results:
#     dest_svc = result['metric']['value']
#     src_svc = result['metric']['source_workload']
#     name = src_svc + '_' + dest_svc
    name = 'er'
    values = result['values']

    values = list(zip(*values))
    if 'timestamp' not in latency_df:
        timestamp = values[0]
        latency_df['timestamp'] = timestamp
        latency_df['timestamp'] = latency_df['timestamp'].astype('datetime64[s]')
    metric = values[1]
    latency_df[name] = pd.Series(metric)
    latency_df[name] = latency_df[name].astype('float64')  * 1000
latency_df
```

```python
import pandas as pd
df = pd.read_csv('out/rate_node_forks_total_52.csv')
print(df.head())

```

```python
from sklearn.cluster import Birch
from sklearn.preprocessing import normalize
import numpy as np
import matplotlib.pyplot as plt
```

```python
smoothing_window = 12
val = df['values']
val = val.rolling(window=smoothing_window, min_periods=1).mean()
x = np.array(val)
normal_x = normalize([x])
norm_x = normal_x.reshape(-1, 1)
# norm_x
plt.figure(figsize=(10, 6))
plt.plot(range(df.shape[0]), norm_x, color='r')
plt.show()
```

```python
threshold = 0.00029
brc = Birch(branching_factor=50, n_clusters=None,threshold=threshold,compute_labels=True)
brc.fit(norm_x)
pred = brc.predict(norm_x)
labels = brc.labels_

plt.figure(figsize=(6, 4))
plt.plot(range(len(labels)), pred, color='r')
plt.show()
```
